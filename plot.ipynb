```python
%matplotlib inline
from mpl_toolkits import mplot3d

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("test.csv", header=0)
# print(training_data_1['x'])

x = df['x']
y1 = df['y1']
y2 = df['y2']

# print(x,y1,y2)
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
# ax1.set_xscale("log")
ax1.set_xlabel("Compensation Constant (comp_const)")
ax1.set_ylabel("Test Perplexity (Test PPL)", color='black')
ax1.plot(x, y1, color=color, label ="Test PPL")
ax1.tick_params(axis='y', labelcolor='black')
# ax1.set_yscale("log")
ax1.axvline(2.7, color='green', ls='-.', linewidth=0.5, ymin=0.02, ymax=0.98, label="2.35")
# ax1.text(1.75, 160, "2.35")
ax1.axvline(2.35, color='orange', ls='-.', linewidth=0.5, ymin=0.02, ymax=0.98, label="2.7")
# ax1.text(2.75, 160, "2.7")
# ax1.axhline(86.60959481, color='blue', ls='-.',linewidth=0.5)


# ax1.axhline(86.060864, color='green', ls='--', label=">=80%")

# ax1.axvline(x_axis[np.argwhere(x.min()], color='green', ls='--', label=">=80%")

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Compression Ratio (CR)', color='black')  # we already handled the x-label with ax1
ax2.plot(x, y2, color=color, label ="CR")
ax2.tick_params(axis='y', labelcolor='black')
# ax2.set_yscale("log")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.legend(bbox_to_anchor=(0.8, 0.95))
plt.savefig("comp_const.pdf")
# plt.grid(which='both', axis='both')
# plt.grid(True)
plt.tight_layout()
plt.show()
# print(x_axis[np.argwhere(np.cumsum(frequencies)>=80)[0]])
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv("test2.csv", header=0)

x = df['x']
y1 = df['y1']
y2 = df['y2']
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
# ax1.set_xscale("log")
ax1.set_xlabel("Clipping Constant (clip_const)")
ax1.set_ylabel("Test Perplexity (Test PPL)", color='black')
ax1.plot(x, y1, color=color, label ="Test PPL")
ax1.tick_params(axis='y', labelcolor='black')
ax1.axvline(60, color='orange', ls='-.', linewidth=0.5, ymin=0.02, ymax=0.98, label="60")

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Compression Ratio (CR)', color='black')  # we already handled the x-label with ax1
ax2.plot(x, y2, color=color, label ="CR")
ax2.tick_params(axis='y', labelcolor='black')
# ax2.set_yscale("log")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.legend(bbox_to_anchor=(0.85, 0.95))
plt.savefig("clip_const.pdf")
# plt.grid(which='both', axis='both')
# plt.grid(True)
plt.tight_layout()
plt.show()
# print(x_axis[np.argwhere(np.cumsum(frequencies)>=80)[0]])
```
